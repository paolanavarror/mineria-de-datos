```python
x1 = np.random.randint(10, size=6)  # One-dimensional array Genera un arreglo aletorio de enteros del 0 al 10 de tamaño 6
x2 = np.random.randint(10, size=(3, 4))  # Two-dimensional array, 3 renglones 4 columnas
x3 = np.random.randint(10, size=(3, 4, 5))  # Three-dimensional array,
```

```python
print(x2.dtype) #Nos dice el tipo de datos
```

```python
print("dtype:", x3.dtype) #los 64 bits estan agrupados en 8 bytes
```

```python
#No es lo mismo almacenar entero que flotantes, se coupa mas memoria en flotantes
print("itemsize:", x3.itemsize, "bytes")
print("nbytes:", x3.nbytes, "bytes") #nbytes
```

Media (Mean)

```python
data = np.array([1, 2, 3, 4, 5])
media = np.mean(data)
print(media)
#se observa que regresa un num flotante
```

Promedio (Average) y Promedio Ponderado

```python
data = np.array([1, 2, 3, 4, 5])
weights = np.array([1, 2, 3, 4, 5])
promedio = np.average(data)
promedio_ponderado = np.average(data, weights=weights)
```

```python
#Lo que hace diferente a promedio a un promedio ponderado a cada elemento le pudede dar un peso diferente
#Promedio ponderado: Para darle un peso distinto a cada uno de los elementos
print(promedio)
print(promedio_ponderado)
```

Mediana

```python
mediana = np.median(data)
```
