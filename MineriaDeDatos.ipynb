```python
# Tablas cruzadas (NumPy)
#definifición de funcipon
def tabla2x2(a, b): #palabra reservada def es para definir una función
    t = np.zeros((2,2), dtype=int) #np.zeros genera un arreglo, una matriz de 2 renglones y de 2 columnas rellenas de 0 de tipo entero
    t[1,1] = np.count_nonzero(a & b) #asigna valores utilizando coordenadas, es decir en la, esta haciendo una op and
    t[1,0] = np.count_nonzero(a & ~b) # not, se eta haciesnd una op and pero se esta negando  el parametro a
    t[0,1] = np.count_nonzero(~a & b) #
    t[0,0] = np.count_nonzero(~a & ~b)
    return t

tab_py_colab = tabla2x2(prog_py, usa_colab)
tab_py_gh    = tabla2x2(prog_py, cuenta_gh)
tab_colab_gh = tabla2x2(usa_colab, cuenta_gh)
```

Primer intento de modelar los años de escolaridad y el ingreso estimado.
Utilizar el primer punto y el útlimo para trazar una recta: $$ m = \frac{y_{2}-y_{1}}{x_{2}-x_{1}} $$
$$ b = y_{1} - m * x_{1} $$

```python
#Para que mi foermula quede centrada se coloca el $$
#para colocar una fracción
```

```python
x1 = anios_escolaridad[0]
y1 = ingreso_mensual_estimado_mx[0]
x2 = anios_escolaridad[-1]
y2 = ingreso_mensual_estimado_mx[-1]

m = (y2 - y1) / (x2 - x1)
b = y1 - m * x1

print(f"La ecuación de la recta es: y = {m:.2f}x + {b:.2f}")
```

```python
linea_ingreso_estimado_new_plot = m * anios_escolaridad + b
linea_ingreso_estimado_new_plot #toma m lo multiplica por años de escolaridad y lo suma a b
```

```python
import matplotlib.pyplot as plt

# Calcular los valores y de la recta utilizando m y b
linea_ingreso_estimado_new_plot = m * anios_escolaridad + b #La ventaja de vectorizar, me permite

plt.figure(figsize=(10, 6))
plt.scatter(anios_escolaridad, ingreso_mensual_estimado_mx, label='Datos Originales')
plt.plot(anios_escolaridad, linea_ingreso_estimado_new_plot, color='red',
         label=f'Recta de Regresión (y = {m:.2f}x + {b:.2f})') #Gráfica
plt.title('Años de Escolaridad vs. Ingreso Mensual Estimado con Línea de Regresión')
plt.xlabel('Años de Escolaridad')
plt.ylabel('Ingreso Mensual Estimado (MXN)')
plt.grid(True)
plt.legend()
plt.show()
```

Valores predichos  (ypred) : Son los ingresos mensuales que nuestra recta de regresión predice para cada año de escolaridad.

Errores (residuales): Es la diferencia entre el ingreso mensual real (ingreso_mensual_estimado_mx) y el ingreso mensual predicho (y_pred). Un error positivo significa que la predicción fue menor que el valor real, y un error negativo indica que la predicción fue mayor.

Errores al cuadrado: Hemos elevado al cuadrado cada uno de los errores. Esto se hace para eliminar los signos negativos (de modo que los errores positivos y negativos no se cancelen entre sí) y para penalizar más los errores grandes.

Suma de errores cuadrados: Es la suma de todos los errores al cuadrado. Este valor nos da una medida total de la 'distancia' de nuestros puntos a la línea.

Error Cuadrático Medio (MSE): Es la suma de los errores cuadrados dividida por el número total de puntos de datos. El MSE es una medida promedio de la magnitud de los errores. Un MSE más bajo indica que el modelo se ajusta mejor a los datos.

En este caso, el Error Cuadrático Medio (MSE) es de aproximadamente 13,196,667.31. Este número nos da una idea del rendimiento general de la recta que trazamos, mostrando la magnitud promedio de los errores al cuadrado entre los valores observados y los valores predichos por la línea.

Fórmula del MSE (Error Cuadrático Medio) del Inglés Mean Squared Error

$$ MSE = \frac {\Sigma_{i=1}^{n}(Y_{i}-\hat{Y_{i}})^{2}}{n} $$
Donde:  
$ MSE = $ error cuadrático medio.    
$ n = $ número de datos.  
$ Y_{i} = $ valores observados.  
$ \hat{Y_{i}} = $ valores predichos.

```python
# 1. Calcular los valores predichos (y_pred) usando la ecuación de la recta
y_pred = m * anios_escolaridad + b
y_pred
```

```python
# 2. Calcular los errores (residuales): la diferencia entre los valores reales y los predichos
#La predicción quedó arriba por eso errores negativos
errores = ingreso_mensual_estimado_mx - y_pred
errores
```

```python
# 3. Elevar al cuadrado cada error
errores_cuadrados = errores**2
errores_cuadrados
```

```python
# 4. Sumar los errores cuadrados
suma_errores_cuadrados = np.sum(errores_cuadrados)
suma_errores_cuadrados
```

```python
# 5. Dividir la suma de los errores cuadrados por el número de puntos para obtener el MSE
mse = suma_errores_cuadrados / len(anios_escolaridad)
mse
```

```python
# Cambia la forma de anios_escolaridad a 2D array como es requerido por sklearn
X = anios_escolaridad.reshape(-1, 1)
y = ingreso_mensual_estimado_mx

X
```

```python
# Crear y ajustar el modelo de regresión lineal
modelo_regresion = LinearRegression()
modelo_regresion.fit(X, y)

# Obtener los coeficientes (m) e intercepto (b) del nuevo modelo
m_sklearn = modelo_regresion.coef_[0]
b_sklearn = modelo_regresion.intercept_

print(f"Coefficients (m) from scikit-learn: {m_sklearn:.2f}")
print(f"Intercept (b) from scikit-learn: {b_sklearn:.2f}")
```

```python
#2.Calculo del MSE de la regresión lineal
#Valores predichos por sklearn ya calculados: y_pred_sklearn
#Calcular errores de sklearn
errores_sklearn = ingreso_mensual_estimado_mx - y_pred_sklearn

#Elevar al cuadrado y promediar
mse_sklearn = np.mean(errores_sklearn**2)

print(f"MSE del primer intento (manual): {mse:,.2f}")
print(f"MSE de la regresión lineal (sklearn): {mse_sklearn:,.2f}")
```

Formulas de los indicadores (markdown)
Indicadores de Evaluación de RegresiónRMSE (Root Mean Squared Error):$$RMSE = \sqrt{\frac{\sum_{i=1}^{n} (Y_i - \hat{Y}_i)^2}{n}}$$MAE (Mean Absolute Error):$$MAE = \frac{1}{n} \sum_{i=1}^{n} |Y_i - \hat{Y}_i|$$$R^2$ (Coeficiente de Determinación):$$R^2 = 1 - \frac{\sum (Y_i - \hat{Y}_i)^2}{\sum (Y_i - \bar{Y})^2}$$MAPE (Mean Absolute Percentage Error):$$MAPE = \frac{100\%}{n} \sum_{i=1}^{n} \left| \frac{Y_i - \hat{Y}_i}{Y_i} \right|$$

```python
import pandas as pd
import numpy as np

np.random.seed(42)
n = 200

df = pd.DataFrame({ #Se esta generadnod un dataframe a partir de un diccionario
"id_cliente": range(1, n+1), #Enteros que empiezan desde el 1 hasta el num n
"genero": np.random.choice(["F", "M", "No binario", "Prefiero no decir"], size=n, p=[0.45, 0.45, 0.05, 0.05]), #el numpy una función aleatoria por ejemplo para una rifa, se esta relacionado al parametro n, el 0.45 esta asociado con las "F" asociada con un 45% y así sucesivamente
"segmento": np.random.choice(["Estándar", "Oro", "Platino"], size=n, p=[0.6, 0.3, 0.1]), #tam de la muestra es n, Estan en orden, la clave es asignarles la probaibilida que sume el 100%
"satisfaccion": np.random.choice([1,2,3,4,5], size=n, p=[0.1,0.2,0.35,0.25,0.1]),
"ciudad": np.random.choice(["SLP", "CDMX", "GDL", "MTY", "QRO"], size=n, p=[0.35,0.25,0.15,0.15,0.10]),
"edad": np.random.normal(35, 10, size=n).round(0).clip(18, 70).astype(int), #Una distribución normal, redondear a 0 es decir enteros,
"ingreso_mensual": np.random.lognormal(mean=9.5, sigma=0.5, size=n).round(2), # pesos MXN aprox. redondeado a 2 decimales
"temperatura_sala": np.random.normal(23, 2, size=n).round(1),
"visitas_ultimo_anio": np.random.poisson(lam=3, size=n), #Distribuci´pon poisson se utiliza en datos pequeños, cuando la muestra es menor a 30
"fecha_ultima_visita": pd.to_datetime("2025-12-31") - pd.to_timedelta(np.random.randint(0, 365, size=n), unit="D") #Metodo para comvnetir fecha y hora
})

df.head() #Nos trae los primeros 5 renglones por defecto
```

```python
df.dtypes #Nos indica el tipo de datos
```

```python
# Nominal → category (sin orden)
df["genero"] = df["genero"].astype("category") #Se cambia el tipo de dato de n
df["ciudad"] = df["ciudad"].astype("category")

# Ordinal → category con orden
segmento_cat = pd.CategoricalDtype(categories=["Estándar", "Oro", "Platino"], ordered=True) #Se puede modelar el tipo de datos en pandas, teniendo un parametro, en Phy una lista es mutable
df["segmento"] = df["segmento"].astype(segmento_cat) #Se configura el tipo de dato y dspues

satisfaccion_cat = pd.CategoricalDtype(categories=[1,2,3,4,5], ordered=True) #Le decimos el orden de las categoria
df["satisfaccion"] = df["satisfaccion"].astype(satisfaccion_cat)

# Intervalo / Razón → numérico (ya están), pero verifica:
df["temperatura_sala"] = pd.to_numeric(df["temperatura_sala"])
df["edad"] = pd.to_numeric(df["edad"])
df["ingreso_mensual"] = pd.to_numeric(df["ingreso_mensual"])
df["visitas_ultimo_anio"] = pd.to_numeric(df["visitas_ultimo_anio"])

# Derivada útil: días desde última visita (razón)
df["dias_desde_ultima_visita"] = (pd.to_datetime("2025-12-31") - df["fecha_ultima_visita"]).dt.days #Tiene una derivada ya que es un dato dinamico
```

Razón (media, mediana, IQR, coeficiente de variación):

```python
def resumen_razon(s): #Las funciones dse definen con las palabras redervadas def, entre parentesisi los parametros
    return pd.Series({ #La fun regresa una serie de pandas
    "n": s.count(),
    "media": s.mean(),
    "mediana": s.median(),
    "sd": s.std(),
    "min": s.min(),
    "q1": s.quantile(0.25),
    "q3": s.quantile(0.75),
    "max": s.max(),
    "cv": s.std()/s.mean() if s.mean()!=0 else np.nan #Coeficiente de variación, cuida la division entre 0, calula esta división simpre y cuando el denominador sea 0  (oneliner de Python)
    }).round(2) #Hasta aquí se define la fun

razon_stats = pd.concat({ #Une de manera horizontal como columna, finalmente regresa un df con esas columnas
    "edad": resumen_razon(df["edad"]), #Genera una serie y luego la mete como una columna en el df
    "ingreso_mensual": resumen_razon(df["ingreso_mensual"]),
    "visitas_ultimo_anio": resumen_razon(df["visitas_ultimo_anio"]),
    "dias_desde_ultima_visita": resumen_razon(df["dias_desde_ultima_visita"])
    }, axis=1) #El eje 0 son los renglones y el eje 1 es el eje de columnas
razon_stats
```
