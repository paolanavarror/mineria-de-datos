#### Conceptos

##### Minería de Datos:
Es una técnica asistida por computadora que se utiliza en los análisis para procesar y explorar grandes conjuntos de datos.
##### Ciencia de datos:
Es el estudio de datos con el fin de extraer información significativa para empresas. Es un enfoque multidisciplinario que combina principios y prácticas del campo de las matemáticas, la estadística, la inteligencia artificial y la ingeniería de computación para analizar grandes cantidades de datos.
