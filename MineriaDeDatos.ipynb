```python
data = np.array([1, 2, 3, 4, 5])
weights = np.array([1, 2, 3, 4, 5])
promedio = np.average(data)
promedio_ponderado = np.average(data, weights=weights)
```

```python
# Descriptivos de edad (NumPy)
summary_age = {
    'n': int(edad.size), #edad es un arreglo del numpy, cuantos elemnetos tiene el arreglo de edad
    'min': float(np.min(edad)), #va a traer la edad menor
    'p25': float(np.percentile(edad, 25)), #Va a traer el 25% de los datos
    'mediana': float(np.median(edad)), #medida de tendencia central, se ordenan los datos de menor a maypr y se obtiene
    'media': float(np.mean(edad)), #promedio
    'p75': float(np.percentile(edad, 75)), #el 75% de los datos que
    'max': float(np.max(edad)),
    'rango(ptp)': float(np.ptp(edad)),
    'var_muestral': float(np.var(edad, ddof=1)), #variancia
    'std_muestral': float(np.std(edad, ddof=1)), #std: desviación estandar
    'cv': float(np.std(edad, ddof=1)/np.mean(edad)), #cv: coeficiente de variancia
}
```

```python
# Tablas cruzadas (NumPy)
#definifición de funcipon
def tabla2x2(a, b): #palabra reservada def es para definir una función
    t = np.zeros((2,2), dtype=int) #np.zeros genera un arreglo, una matriz de 2 renglones y de 2 columnas rellenas de 0 de tipo entero
    t[1,1] = np.count_nonzero(a & b) #asigna valores utilizando coordenadas, es decir en la, esta haciendo una op and
    t[1,0] = np.count_nonzero(a & ~b) # not, se eta haciesnd una op and pero se esta negando  el parametro a
    t[0,1] = np.count_nonzero(~a & b) #
    t[0,0] = np.count_nonzero(~a & ~b)
    return t

tab_py_colab = tabla2x2(prog_py, usa_colab)
tab_py_gh    = tabla2x2(prog_py, cuenta_gh)
tab_colab_gh = tabla2x2(usa_colab, cuenta_gh)
```
